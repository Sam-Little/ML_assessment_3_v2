Exploratory Data Analysis (EDA)

In this section I will explore different relationsships between the features and will use visulisations to understand.

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 

# Load the processed dataset
data = pd.read_csv(r'C:\Users\Admin\ML_assessment_3_v2\data\processed\processed_data.csv')

# Boxplot for Parental Involvement vs Exam Score
sns.boxplot(x='Parental_Involvement', y='Exam_Score', data=data)

# Add title and labels
plt.title("Impact of Parental Involvement on Exam Scores")
plt.xlabel("Parental Involvement")
plt.ylabel("Exam Score")

# Display the plot
plt.show()
```
